```python
ADF = pd.DataFrame({
    'name':['Sherry','Hamza','Zulqarnain'],
    'Class':[np.nan,11,12],
    'born':['26-2-2004',np.nan,np.nan]
})
ADF
```

```python
ADF.dropna(how='all', axis=1) 
# if there are all values are none in column dropna will remove that column 
```

```python
ADF.dropna()
# if there are any of the value in a row is none this will remove the whole row
```
